```python
import os
import cv2
import pandas as pd

path = "/content/gdrive/MyDrive/Dataset/input/frames_rgb"

marked = {0 : set(),
          1: set(),
          2 : set(),
          3 : set()
}
columns = ['frame_id', 'wood', 'glass', 'plastic', 'metal']
df = pd.DataFrame(columns=columns)
#os.path.join(path, filename)
for filename in sorted(os.listdir(path)):
        results = model.track(cv2.imread(os.path.join(path, filename)),persist=True, verbose=False)
        local = {0 : set(),
          1: set(),
          2 : set(),
          3 : set()
        }
        for res in results[0].boxes.data:
            cl = int(res[6].item())
            local[cl].add(int(res[4].item()))
            marked[cl].add(int(res[4].item()))
            # df = df.append(pd.Series(['f' + str(len(df)),len(local[0]), len(local[1]), len(local[2]), len(local[3])], index=df.columns[:5)]), ignore_index=True)
        df.loc[len(df), :] = ['f' + str(len(df)),len(local[0]), len(local[1]), len(local[2]), len(local[3])]
df.loc[len(df)] = ['f_all' ,len(marked[0]), len(marked[1]), len(marked[2]), len(marked[3])]
df.to_csv("submission.csv", sep = ';', index = False)

```
